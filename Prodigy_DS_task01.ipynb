```python
Task-1
Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable,such as the distribuion of
ages or genders in a population.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# continuous variable data
data = np.random.randn(2000)

# Create a histogram
plt.hist(data, bins=40, color='lightgreen', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Continuous Variable')
plt.show()
```
